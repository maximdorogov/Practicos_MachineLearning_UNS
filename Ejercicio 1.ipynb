## Importo las librerias que voy a usar

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url_dataset_A = 'https://raw.githubusercontent.com/manlio99/Materia-de-aprendizaje/master/4_DataWrangling/data/camas_covid_bahia.csv'
url_dataset_B = 'https://raw.githubusercontent.com/manlio99/Materia-de-aprendizaje/master/4_DataWrangling/data/casos_covid_bahia.csv'

data_frame_A = pd.read_csv(url_dataset_A)
data_frame_B = pd.read_csv(url_dataset_B)
```

## Imprimo el histograma de algunos campos

# Itero sobre las columnas y grafico el histograma

## Podemos ver como los campos de "respiradores ocupados" o "camas sospechosos covid" siguen una distribucion que se podria aproximar a gausiana.

```python
for key in data_frame_A.keys():
    if(key != 'fecha' and key != 'observaciones'): data_frame_A.hist(key)

```

```python
for key in data_frame_B.keys():
    if(key != 'fecha'): data_frame_B.hist(key)
```

## En la diagonal principal del grafico que se muestra a continuacion vemos la distribucion marginal de las variables que conforman el dataset A. Algunas con la  forma de la tipica campana de gaus

```python
sns.pairplot(data_frame_A, diag_kind="kde")
```
